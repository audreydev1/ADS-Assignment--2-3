# 1.3 Analysis

## 1.3.1 Uncovering Factors that led to Employee Attrition

```python
#Finding out how many employees attrited from the company
EmployeeAttrition["Attrition"].value_counts()
```

```python
#subsetting our data set to only have employees who have attrited.
att_Yes=EmployeeAttrition[EmployeeAttrition["Attrition"]=="Yes"]
att_Yes.shape
```

```python
#plotting this data to get a clear view of how many employees attrited
plt.figure(figsize=(8,8))
sns.set_context("paper",font_scale=1.5)
ax=sns.countplot(x="Attrition",data=EmployeeAttrition)
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")

plt.title("Employee Attrition")
plt.show()
```

### From the analysis we can observe that 237 employees attrited out of the 1470 employees,which is about 16.12%

## 1.3.1.1 Age with Respect to Attrition

```python
#Analysing the Age Column
#Checking on how employee ages are distributed
plt.figure(figsize=(8,8))
sns.set_context("paper",font_scale=1.5)
sns.histplot(data=EmployeeAttrition, x="Age", hue="Attrition", multiple="stack",palette="magma",bins=np.arange(20,70,5))
plt.title("Hisogram for Employees Age Distribution by Attrition")
plt.show()
```

### Most of the employees fall in the age bracket of between 30 years to 40 years,and we can also observe that attrition is high in the early ages as to compared to older ages.

```python
# Checking for attrition across different ages
plt.figure(figsize=(26,20))
sns.set_context("paper",font_scale=2)
ax=sns.histplot(x="Age",hue="Attrition",data=att_Yes,palette="magma",bins=np.arange(20,70,5))
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")
    
plt.ylabel("No. of employees that Attrited")    
plt.title("Histogram showing the Distribution of Attrited Employees")
plt.show()
```

### From our visualization we can observe that employees in ages between 25 to 35  tend to attrite more

```python
#Attrition by age and business travel
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
ax=sns.barplot(x="BusinessTravel",y="Age",hue="Attrition",data=EmployeeAttrition,palette="BuGn",ci=None)
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")
    
plt.title("Employee Attrition with respect to Age and Business Travel")
plt.show()
```

## 1.4.1 Breakdown of distance from home by job role and attrition

```python
#Checking on how the distance from home is distributed
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
ax=sns.histplot(data=EmployeeAttrition, x="DistanceFromHome", hue="Attrition", multiple="stack",palette="magma",bins=np.arange(0,30,5))

plt.title("Hisogram for Attrition with respect to Distance From Home ")
plt.ylabel("Number of Employees")
plt.show()
```

### We can observe that the distance from home for most employees is between 0-10 Miles

```python
# Summary Statistics for the distance from home column from those who attrited
att_Yes['DistanceFromHome'].describe()
```

```python
#Attrition with respect to Distance From Home and Job Role
EmployeeAttrition.groupby(['JobRole', 'Attrition']).agg({'DistanceFromHome':np.mean})
```

```python
#plotting the data
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
ax=sns.barplot(x="JobRole",y="DistanceFromHome",hue="Attrition",data=EmployeeAttrition,palette="viridis",ci=None)
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")
plt.xticks(rotation=45)    
plt.title("Employee Attrition by Job Role and Distance From Home")
plt.show()
```

###  We can observe that employees who stay more than 9 kilometers from the office are more likely to attrite regardless of the job role

## 1.4.2 Comparing Average Monthly Income By Education and Attrition

```python
#Summary Statistics For the monthly income Statistics
EmployeeAttrition["MonthlyIncome"].describe()
```

```python
#checking the distribution of monthly income
plt.figure(figsize=(8,8))
sns.set_context("paper",font_scale=1.5)
ax=sns.histplot(data=EmployeeAttrition, x="MonthlyIncome", hue="Attrition", multiple="stack",palette="magma")
plt.title("Histogram for Monthly Income")
plt.show()
```

```python
#analysing the education column
EmployeeAttrition["Education"].value_counts()
```

```python
#checking how Education and job role are distributed
pd.crosstab(EmployeeAttrition['Education'], EmployeeAttrition['JobRole'], normalize = 'index')
```

```python
#plotting the data for employee attrition by Education
plt.figure(figsize=(16,10))
sns.set_style("dark")
sns.set_context("paper",font_scale=1.5)
ax=sns.countplot(x="Education",hue="Attrition",data=EmployeeAttrition,palette="summer")
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")
    
plt.title("Employee Attrition by Education")
plt.show()
```

### Employees with education level 3 are the majority in the organization and they tend to attrite more.

```python
#compairing average monthly income by education and attrition
EmployeeAttrition.groupby(['Education', 'Attrition']).agg({'MonthlyIncome':np.mean})
```

```python
#plotting the data
plt.figure(figsize=(16,10))
sns.set_context("paper",font_scale=1.5)
ax=sns.barplot(x="Education",y="MonthlyIncome",hue="Attrition",data=EmployeeAttrition,palette="viridis",ci=None)
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        height=0
    ax.text(p.get_x()+p.get_width()/2., height, '{:1.2f}'.format(height), ha="center")   
plt.title("Average Monthly Income by Attrition and Education")
plt.show()
```

### We can observe that the average monthly income increases with the education level i.e. the higher the level of education the higher the income.
### Average monthly income for those who attrited is less than those who have not attrited across all the education levels.
